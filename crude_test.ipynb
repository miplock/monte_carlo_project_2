sprawdzamy czy funkcje ktore stworzylem dzialają

```python
from methods import (crude_monte_carlo, control_variate_monte_carlo,
                     stratified_monte_carlo, antithetic_monte_carlo)
from parameters import S0, K, r, seed, sigma, mu_star, T, R, n_steps
```

```python
C = 105        # np. jedna wartość bariery na początek

est, se, var = crude_monte_carlo(S0, mu_star, sigma, r, K,
                                 C, T, n_steps, R, seed)
print("Crude MC estimate:", est)
print("Std error:", se)
```

```python
est, se, var = control_variate_monte_carlo(S0, mu_star, sigma, r, K,
                                 C, T, n_steps, R, seed)
print("control variate MC estimate:", est)
print("Std error:", se)
```

```python
est, se, var = stratified_monte_carlo(S0, mu_star, sigma, r, K,
                                 C, T, n_steps, R, seed)
print("stratified MC estimate:", est)
print("Std error:", se)
```

```python
est, se, var = antithetic_monte_carlo(S0, mu_star, sigma, r, K,
                                 C, T, n_steps, R if R % 2 == 0 else R + 1, seed)
print("antithetic MC estimate:", est)
print("Std error:", se)
```
