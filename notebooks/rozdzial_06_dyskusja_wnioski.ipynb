# 6. Dyskusja i wnioski końcowe

Przeprowadzone symulacje Monte Carlo umożliwiły szczegółowe porównanie różnych estymatorów dla opcji europejskiej oraz dyskretnej opcji barierowej typu up-and-out. Analiza tych dwóch przypadków pokazuje, że skuteczność metod Monte Carlo oraz technik redukcji wariancji w istotnym stopniu zależy od struktury funkcji wypłaty oraz charakteru zdarzeń determinujących wartość instrumentu.

W przypadku opcji europejskiej, dla której funkcja wypłaty jest gładka i nie zależy od całej trajektorii procesu, wszystkie rozważane estymatory zachowywały się zgodnie z przewidywaniami teoretycznymi. Techniki redukcji wariancji prowadziły do systematycznego zmniejszenia wariancji estymatora, a najlepsze wyniki uzyskano przy użyciu estymatora z wariantem kontrolnym. Próbkowanie warstwowe z alokacją optymalną było wyraźnie skuteczniejsze od alokacji proporcjonalnej, natomiast metoda antytetyczna zapewniała stabilną, choć umiarkowaną poprawę względem estymatora naiwniego.

Analiza opcji barierowej ujawniła znacznie bardziej złożone zachowanie estymatorów. Dla relatywnie wysokiego poziomu bariery ($C=120$) wszystkie metody dawały porównywalne wyniki, a różnice w wariancji miały charakter drugorzędny. W tym przypadku wpływ warunku barierowego był ograniczony, a większość trajektorii zachowywała się podobnie jak w przypadku opcji europejskiej.

Sytuacja istotnie zmieniała się dla bariery położonej blisko ceny początkowej aktywa ($C=105$). W tym przypadku wypłata opcji była zdominowana przez rzadkie zdarzenia, w których proces nie przekracza bariery przez cały horyzont czasowy i jednocześnie kończy się powyżej ceny wykonania. Prowadziło to do znacznego wzrostu wariancji estymatorów oraz ujawniło istotne ograniczenia niektórych technik redukcji wariancji.

W szczególności próbkowanie warstwowe z alokacją optymalną okazało się w tym przypadku niestabilne. Obserwowano silnie asymetryczne rozkłady błędów, liczne obserwacje odstające oraz drastyczne obniżenie empirycznego pokrycia przedziałów ufności. Zjawisko to można tłumaczyć faktem, że alokacja optymalna opiera się na estymacji wariancji w poszczególnych warstwach, która w przypadku silnie nieliniowej i zdyskontynuowanej funkcji wypłaty może być bardzo niedokładna. W konsekwencji metoda, która w teorii minimalizuje wariancję, w praktyce prowadziła do pogorszenia jakości estymacji.

Warto również podkreślić, że rozważana opcja barierowa ma charakter dyskretny, tzn. warunek barierowy sprawdzany jest jedynie w skończonej liczbie punktów czasowych. Oznacza to, że estymowana cena odpowiada dyskretnej opcji barierowej, a nie jej ciągłemu odpowiednikowi. Taka dyskretyzacja prowadzi do systematycznego zawyżenia wartości opcji w porównaniu z barierą ciągłą, gdyż potencjalne przekroczenia bariery pomiędzy punktami dyskretyzacji nie są rejestrowane.

Uzyskane wyniki pokazują, że w przypadku opcji o wypłatach zależnych od całej trajektorii procesu stabilność estymatora oraz wiarygodność estymacji niepewności są często ważniejsze niż teoretyczna minimalizacja wariancji. Proste i stabilne metody, takie jak estymator naiwny lub próbkowanie warstwowe z alokacją proporcjonalną, mogą w praktyce okazać się bardziej niezawodne niż metody teoretycznie optymalne.

Naturalnym kierunkiem dalszych badań byłoby zastosowanie metod zaprojektowanych specjalnie do estymacji rzadkich zdarzeń. W szczególności można rozważyć importance sampling, polegający na modyfikacji rozkładu przyrostów procesu w taki sposób, aby zwiększyć częstość trajektorii nieprzekraczających bariery, przy jednoczesnym zachowaniu nieobciążoności estymatora. Innym podejściem są metody typu splitting lub subset simulation, w których zdarzenie nieprzekroczenia bariery jest osiągane stopniowo poprzez sekwencję warunków pośrednich.

Dodatkowym kierunkiem analizy mogłoby być porównanie dyskretnej i ciągłej wersji opcji barierowej, na przykład poprzez zastosowanie korekty Browna (Brownian bridge) do wykrywania przekroczeń bariery pomiędzy punktami dyskretyzacji. Pozwoliłoby to ocenić wpływ schematu czasowego na estymowaną cenę opcji oraz jakość estymacji Monte Carlo.

Przedstawione wyniki potwierdzają, że metody Monte Carlo stanowią elastyczne i potężne narzędzie do wyceny złożonych instrumentów finansowych, jednak ich skuteczność w dużym stopniu zależy od właściwego doboru technik symulacyjnych do specyfiki analizowanego problemu.
