## Estymator stratified z alokacją optymalną (Neymana)

Korzystamy z konstrukcji warstw $A_1,\dots,A_m$ w przestrzeni
$Z\sim\mathcal N(0,I_n)$ z sekcji 2.2 PDF, gdzie
$\mathbb P(Z\in A_i)=p_i=1/m$, oraz z faktu, że trajektorię ruchu Browna
$B=(B(1/n),\dots,B(1))$ otrzymujemy przez $B=AZ$.

Niech
$$
Y = A_{n,C}
$$
oznacza zdyskontowany payoff z (3.3).

### Estymator warstwowy

Dla ustalonej alokacji $R_1,\dots,R_m$ ($\sum_i R_i=R$) generujemy
$Y_{i,1},\dots,Y_{i,R_i}$ i definiujemy
$$
\bar Y_i = \frac{1}{R_i}\sum_{k=1}^{R_i} Y_{i,k}.
$$
Estymator ma postać
$$
\hat I_{\text{strat}} = \sum_{i=1}^m p_i\,\bar Y_i,
\qquad p_i=\frac{1}{m}.
$$

Oszacowanie wariancji:
$$
\widehat{\mathrm{Var}}(\hat I_{\text{strat}})
= \sum_{i=1}^m \frac{p_i^2\, s_i^2}{R_i},
\qquad
s_i^2=\frac{1}{R_i-1}\sum_{k=1}^{R_i}(Y_{i,k}-\bar Y_i)^2.
$$

### Alokacja optymalna (Neymana)

Wariancja estymatora jest minimalizowana przez wybór
$$
R_i^* = R \cdot
\frac{p_i\,\sigma_i}{\sum_{j=1}^m p_j\,\sigma_j},
$$
gdzie $\sigma_i^2=\mathrm{Var}(Y\mid Z\in A_i)$.
Ponieważ $p_i=1/m$, otrzymujemy
$$
R_i^* = R \cdot
\frac{\sigma_i}{\sum_{j=1}^m \sigma_j}.
$$

### Implementacja alokacji optymalnej

Nie znamy $\sigma_i$, dlatego:

1. Wykonujemy **pilot run**: w każdej warstwie $i$ generujemy $r_0$ prób
   i estymujemy $\hat\sigma_i$.
2. Wyznaczamy alokację:
   $$
   \hat R_i = R \cdot
   \frac{p_i\,\hat\sigma_i}{\sum_{j=1}^m p_j\,\hat\sigma_j}
   \quad
   (\text{tu: } \hat R_i = R\cdot
   \tfrac{\hat\sigma_i}{\sum_j \hat\sigma_j}).
   $$
3. Po zaokrągleniu $\hat R_i$ wykonujemy właściwą symulację i liczymy
   $\hat I_{\text{strat}}$ oraz
   $\widehat{\mathrm{Var}}(\hat I_{\text{strat}})$ ze wzoru warstwowego.

Tak skonstruowany estymator wykorzystuje stratyfikację z sekcji 2.2 PDF
i minimalizuje wariancję przy zadanym budżecie symulacji $R$.
